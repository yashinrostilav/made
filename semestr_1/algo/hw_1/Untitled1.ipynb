```python
def heapsort(arr, n):
    heap(arr, n)
    for i in range(n - 1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]
        heapify(arr, 0, i)
 
def parent(i):
    return (i - 1) // 2
 
def left(i):
    return 2 * i + 1 
 
def heap(arr, n):
    
    start = parent(n - 1)
    while start >= 0:
        heapify(arr, start, n)
        start -= 1
 
def heapify(arr, index, n):
    l = left(index)
    r = l+1
    if (l < n and arr[l] > arr[index]):
        largest = l
    else:
        largest = index
    if (r < n and arr[r] > arr[largest]):
        largest = r
    if (largest != index):
        arr[largest], arr[index] = arr[index], arr[largest]
        heapify(arr, largest, n)
 
arr = []
n = int(input())
arr = list(map(int,input().strip().split()))[:n]
        
array = heapsort(arr, n)
ans = ' '.join(map(str, arr))
print (ans)
```

```python
def qsort(arr, left, right):
    if (right > left):
        pivot_point = arr[right]
        low = left - 1
        high = right
        
        while (True):
            low = low + 1
            high = high - 1                            
            while(arr[low] < pivot_point):
                low = low + 1
            while(arr[high] > pivot_point):
                high = high - 1
            if (low >= high):
                break;
            arr[low], arr[high] = arr[high], arr[low]
        arr[low], arr[right] = arr[right], arr[low]
        
        qsort(arr, left, low - 1)
        qsort(arr, low + 1, right)
        
arr = []
n = int(input())
arr = list(map(int,input().strip().split()))[:n]
        
qsort(arr, 0, n - 1)
ans = ' '.join(map(str, arr))
print (ans)
```
